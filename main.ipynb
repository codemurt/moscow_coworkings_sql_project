```python
# Where с объединением трёх таблиц в where 
query = """
SELECT full_name FROM coworkings 
WHERE coworking_id IN (
    SELECT coworking_id FROM locations 
    WHERE adm_area = 'Восточный административный округ'
)
AND coworking_id IN (
    SELECT coworking_id FROM tariffs 
    WHERE tariff_type = 'daily' AND price <= 1000
)
AND coworking_id IN (
    SELECT coworking_id FROM facilities 
    WHERE facility_name = 'Wi-Fi'
)
"""
pd.read_sql(query, connection)
```

```python
# 3.1 Простой запрос
query = """
SELECT full_name, avg_rating FROM coworkings 
WHERE avg_rating > (SELECT AVG(avg_rating) FROM coworkings)
"""
pd.read_sql(query, connection)
```

```python
# 3.2 Невозможно переписать
query = """
SELECT full_name, (
    SELECT COUNT(*) 
    FROM facilities f 
    WHERE f.coworking_id = c.coworking_id
) AS facilities_count 
FROM coworkings c
"""
pd.read_sql(query, connection)
```

```python
# 3.3 WITH (CTE)
query = """
WITH TopRated AS (
    SELECT coworking_id 
    FROM coworkings 
    WHERE avg_rating >= 4.0
)
SELECT coworking_id, location_id, adm_area, district, address FROM TopRated 
JOIN locations USING(coworking_id)
"""
pd.read_sql(query, connection)
```

```python
# INNER JOIN
query = """
SELECT c.full_name, t.tariff_type, t.price 
FROM coworkings c
JOIN tariffs t ON c.coworking_id = t.coworking_id
"""
pd.read_sql(query, connection)
```

```python
# LEFT JOIN
query = """
SELECT c.full_name, COUNT(r.review_id) AS reviews_count 
FROM coworkings c
LEFT JOIN reviews r ON c.coworking_id = r.coworking_id 
GROUP BY c.coworking_id
"""
pd.read_sql(query, connection)
```

```python
# Агрегация
query = """
SELECT 
    l.adm_area,
    AVG(c.avg_rating) AS avg_rating,
    COUNT(DISTINCT c.coworking_id) AS coworkings_count
FROM locations l
JOIN coworkings c USING(coworking_id)
WHERE l.district IN ('Измайлово', 'Нагорный район', 'Строгино')
GROUP BY l.adm_area
HAVING avg_rating > 4.0
ORDER BY avg_rating DESC
"""
pd.read_sql(query, connection)
```

```python
# Оконная функция
query = """
SELECT 
    full_name,
    adm_area,
    avg_rating,
    RANK() OVER(PARTITION BY adm_area ORDER BY avg_rating DESC) AS area_rank
FROM coworkings
JOIN locations USING(coworking_id)
"""
pd.read_sql(query, connection)
```

```python
# За рамками курса (использование геофункций):
query = """
SELECT full_name, 
       ST_AsText(coordinates) AS coords,
       ST_Distance_Sphere(
           coordinates, 
           POINT(37.617635, 55.755814)
       ) AS distance_from_kremlin
FROM locations
JOIN coworkings USING(coworking_id)
"""
pd.read_sql(query, connection)
```

```python
# скрипт на языке программирования с обращением к БД

def get_top_coworkings(limit=5):
    query = f"""
    SELECT c.full_name, c.avg_rating, l.address 
    FROM coworkings c
    JOIN locations l USING(coworking_id)
    ORDER BY avg_rating DESC 
    LIMIT {limit}
    """
    return pd.read_sql(query, connection)

print("Топ коворкингов Москвы:")
get_top_coworkings()
```
